```python
import time
import json
import pandas as pd

#Converting to JSON one million entries at a time
with open('review.json') as fp:
    my_line = fp.readline()
    cnt = 1   
    head = ""

    while my_line and cnt < 1000000:
        head = head + my_line 
        my_line = fp.readline()
        cnt += 1

        
    print("1 million done")
    time.sleep(120) #Sleeping for 2 minutes to ensure program doesn't crash
    df = pd.read_json(head, lines = True)

    
    head_2 = ""
    while my_line and cnt < 2000000:
        head_2 = head_2 + my_line
        my_line = fp.readline()
        cnt += 1
        
    print("2 million done")
    time.sleep(120)
    df2 = pd.read_json(head_2, lines = True)
    df = df.append(df2)

    
    head_3 = ""
    while my_line and cnt < 3000000:
        head_3 = head_3 + my_line
        my_line = fp.readline()
        cnt += 1
        
    print("3 million done")
    time.sleep(120)
    df3 = pd.read_json(head_3, lines = True)
    df = df.append(df3)
    
    
    head_4 = ""
    while my_line and cnt < 4000000:
        head_4 = head_4 + my_line
        my_line = fp.readline()
        cnt += 1
        
    print("4 million done")
    time.sleep(120)
    df4 = pd.read_json(head_4, lines = True)
    df = df.append(df4)
    
    
    head_5 = ""
    while my_line and cnt < 5000000:
        head_5 = head_5 + my_line
        my_line = fp.readline()
        cnt += 1
        
    print("5 million done")
    time.sleep(120)
    df5 = pd.read_json(head_5, lines = True)
    df = df.append(df5)
    
    
    head_6 = ""
    while my_line and cnt < 6000000:
        head_6 = head_6 + my_line
        my_line = fp.readline()
        cnt += 1
        
    print("6 million done")
    time.sleep(120)
    df6 = pd.read_json(head_6, lines = True)
    df = df.append(df6)

i = 0
for row in df['text']:
    index = i
    new_row = row.replace(",", "")
    df.at[index, 'text'] = new_row
    i = i + 1
    
print(df.head(100))

df.to_csv('review_without_commas.csv')
print("done")

print(df.head(100))

```
